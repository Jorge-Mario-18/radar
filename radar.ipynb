Stats Radar for FC United vs Sporting Football

```python
from mplsoccer import Radar, FontManager, grid
import matplotlib.pyplot as plt
```

We're gonna set that parameters for the stats we're measuring during the game

```python
# parameter names of the statistics we want to show
params = ["Match Possession %", "1st Half Possession %", "2nd Half Possession %", "Total Shots", "On Target Shots",
          "Off Target Shots", "Shot Accuracy %"]

# The lower and upper boundaries for the statistics
low =  [0, 0, 0, 0, 0, 0, 0]
high = [100, 100, 100, 20, 20, 20, 100]
```

```python
# Instantiate the Radar Class
radar = Radar(params, low, high, round_int=[False]*len(params), num_rings=4,  
              ring_width=1, center_circle_radius=1)
```

```python
# Team Values from the match
FCU =   [52, 56, 48, 15, 3, 12, 41]
SF =  [48, 44, 52, 7, 6, 1, 86]
```

```python
# Create the radar visual
fig, axs = grid(figheight=14, grid_height=0.915, title_height=0.06, endnote_height=0.025,
                title_space=0, endnote_space=0, grid_key='radar', axis=False)

# plot radar
radar.setup_axis(ax=axs['radar'])  # format axis as a radar
rings_inner = radar.draw_circles(ax=axs['radar'], facecolor='#ffb2b2', edgecolor='#fc5f5f')
radar_output = radar.draw_radar_compare(FCU, SF, ax=axs['radar'],
                                        kwargs_radar={'facecolor': '#00f2c1', 'alpha': 0.6},
                                        kwargs_compare={'facecolor': '#d80499', 'alpha': 0.6})
radar_poly, radar_poly2, vertices1, vertices2 = radar_output
range_labels = radar.draw_range_labels(ax=axs['radar'], fontsize=25)
param_labels = radar.draw_param_labels(ax=axs['radar'], fontsize=25)
axs['radar'].scatter(vertices1[:, 0], vertices1[:, 1],
                     c='#00f2c1', edgecolors='#6d6c6d', marker='o', s=150, zorder=2)
axs['radar'].scatter(vertices2[:, 0], vertices2[:, 1],
                     c='#d80499', edgecolors='#6d6c6d', marker='o', s=150, zorder=2)

# Adds the names for the team in the corresponding color 
title1_text = axs['title'].text(0.32, 0.65, 'FC United', fontsize=25, color='#01c49d', ha='left', va='center')
title3_text = axs['title'].text(0.75, 0.65, 'Sporting Football', fontsize=25, ha='right', va='center', color='#d80499')
# Adds the gridlines 
lines = radar.spoke(ax=axs['radar'], color='#a6a4a1', linestyle='--', zorder=2)
```

This was a quick visual of the basic match stats for the game between FC United and Sporting Football. We can see that even though FC United had more shots taken, Sporting Football had a more effective shot accuracy.
